```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as  plt
import seaborn as sns
import datetime as datetime
```

```python
data=pd.read_csv("householdtask3.csv")
```

```python
display(data.head(10))
```

```python
#scatter plot with  year against own
plt.scatter(data["year"],data["own"])
plt.xlabel("year")
plt.ylabel("own")
plt.title("scatter plot")
plt.show()
```

```python
# scatter age against size
plt.scatter(data["age"],data["size"])
plt.xlabel("age")
plt.ylabel("size")
plt.title("scatter plot")
plt.show()
```

```python
#line chart year against own
plt.plot(data["year"])
plt.plot(data["own"])
plt.xlabel("year")
plt.ylabel("own")
plt.title("line plot")
plt.show()
```

```python
# line chart age against size
plt.plot(data["age"])
plt.plot(data["size"])
plt.xlabel("year")
plt.ylabel("own")
plt.title("line plot")
plt.show()
```

```python
#bar graph age against size
plt.bar(data["age"],data["size"])
plt.xlabel("age")
plt.ylabel("size")
plt.title("bar graph")
plt.show()
```

```python
#bar graph age against own
plt.bar(data["age"],data["own"])
plt.xlabel("age")
plt.ylabel("own")
plt.title("bar graph")
plt.show()
```

```python
#histogram of income
plt.hist(data["income"])
plt.title("histogram")
plt.show()
```

```python
#histogram of age
plt.hist(data["age"])
plt.title("histogram")
plt.show()
```
